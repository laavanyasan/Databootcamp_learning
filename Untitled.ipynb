```python
def calc_mean_time(df, category, continent, verbose = False):
    
    mask_asia = df['continent'] == continent
    #mask_asia.head()
    df_asia = df[mask_asia]
    #df_asia.head(10)

    years = df_asia['year'].unique()
    summary = []

#Looping over years to calc the stats needed. 
    for year in years:
        mask_year = df_asia['year'] == year
        df_year = df_asia[mask_year]
        value = np.mean(df_year[category])
        summary.append((continent, year, value))


    results_df = pd.DataFrame(summary, columns = ['continent', 'year', category])
    return results_df
```

```python
calc_mean_time(df, 'lifeexp', 'asia')
```

```python
#tO MAKE THE above more automated : 

def calc_fun_time(df, category, continent,func, verbose = True):
    """"
    
    This is a neat thing you can do with functions in Jupyter!!!
    Here you would mention what the arguements are and why they would be needed
    
    Args : 
    List:
    
    
    Returns:
    also data.
    
    
    """
    
    #Here you can check things to make sure errant arguements don't break your shizz
    assert category in df.columns.values
    assert 'continent' in df.columns.values, 'continent column not found in df'
    #assert continent in df['continent']
    
    
    mask_asia = df['continent'] == continent
    #mask_asia.head()
    df_asia = df[mask_asia]
    #df_asia.head(10)

    years = df_asia['year'].unique()
    summary = []

#Looping over years to calc the stats needed. 
    for year in years:
        mask_year = df_asia['year'] == year
        df_year = df_asia[mask_year]
        value = func(df_year[category])
        summary.append((continent, year, value))


    results_df = pd.DataFrame(summary, columns = ['continent', 'year', category])
    return results_df
```

```python
calc_fun_time(df, 'lifeexp', 'asia', np.median)
```
